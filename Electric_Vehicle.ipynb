```python
evdata = evdata.drop('Legislative District', axis=1)
```

```python
evdata.isnull().sum()
```

```python
evdata.to_csv('EV_Washington_Modified.csv', index=False)
```

```python
evdata.head()
```

```python
evdata_make = evdata.groupby(['State', 'City', 'County', 'Make'])['Make'].count().reset_index(name = 'Make_Count')
evdata_evt = evdata.groupby(['State', 'City','County', 'Electric Vehicle Type'])['Electric Vehicle Type'].count().reset_index(name = 'EVT_Count')
```

```python

evdata_make.to_excel('EV_Washington_Modified_Totalcounts.xlsx', sheet_name='make_data')
```

```python
evdata_evt.to_excel('EV_Washington_Modified_Totalcounts.xlsx', sheet_name='evt_data')
```

```python
with pd.ExcelWriter('EV_Washington_Modified_Totalcounts.xlsx', engine='openpyxl') as writer:
    evdata_make.to_excel(writer, sheet_name='make_data')
    evdata_evt.to_excel(writer, sheet_name='evt_data')
```
